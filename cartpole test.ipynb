```python
import gymnasium as gym
import stable_baselines3 as sb

from dqn import DQN
```

```python
env = gym.make("CartPole-v1")
model = sb.DQN('MlpPolicy', env, verbose=1, learning_rate=1e-3, buffer_size=50000, learning_starts=500, batch_size=64, gamma=0.99, target_update_interval=500)
model.learn(total_timesteps=20000)

env = gym.make("CartPole-v1", render_mode="human")
obs, _ = env.reset()
for i in range(100):
    action, _states = model.predict(obs)
    obs, rewards, terminated, finished, _ = env.step(action)
    env.render()
    if terminated or finished:
        obs, _ = env.reset()

env.close()
```

```python
env = gym.make("CartPole-v1")
model = DQN(env, lr=5e-4)
model.learn()

model.policy.eval()
env = gym.make("CartPole-v1", render_mode="human")
obs, _ = env.reset()
for i in range(100):
    action = model.predict(obs)
    obs, rewards, terminated, finished, info = env.step(action)
    env.render()
    if terminated or finished:
        obs, info = env.reset()

env.close()
```
