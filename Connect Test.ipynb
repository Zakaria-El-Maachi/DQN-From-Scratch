```python
import stable_baselines3 as sb

from dqn import DQN
from qnetconnect import QConnect, ConnectTrainer
from connect import MiniMax
import numpy as np
```

```python
env = ConnectTrainer()
model = sb.DQN('MlpPolicy', env, verbose=1, learning_rate=1e-3, buffer_size=50000, learning_starts=500, batch_size=64, gamma=0.99, target_update_interval=500)
model.learn(total_timesteps=100000)

env = ConnectTrainer(opponent=MiniMax)
obs, _ = env.reset()
r = []
for i in range(1000):
    action, _states = model.predict(obs)
    obs, rewards, terminated, finished, _ = env.step(action)
    r.append(rewards)
    if terminated or finished:
        obs, _ = env.reset()

env.close()
print(np.mean(r))
```

```python
env = ConnectTrainer()
model = DQN(env, lr=1e-3, timesteps=100000, policyNetwork=QConnect)
model.learn()

model.policy.eval()
env = ConnectTrainer(opponent=MiniMax)
obs, _ = env.reset()
r = []
for i in range(1000):
    action = model.predict(obs)
    obs, rewards, terminated, finished, _ = env.step(action)
    r.append(rewards)
    if terminated or finished:
        obs, _ = env.reset()

env.close()
print(np.mean(r))
```
